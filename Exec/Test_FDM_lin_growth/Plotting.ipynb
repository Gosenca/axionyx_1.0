## Time evolution of soliton

```python
data_high = np.transpose(np.loadtxt('runlog_high', skiprows=1))
data_medium = np.transpose(np.loadtxt('runlog_medium', skiprows=1))
data_low = np.transpose(np.loadtxt('runlog_low', skiprows=1))

nstep_high = data_high[0]
time_high = data_high[1]
dt_high = data_high[2]
z_high = data_high[3]
mass_high = data_high[4]/data_high[4][0]
epot_high = data_high[5]/data_high[5][0]
ekinrho_high = data_high[6]/data_high[6][0]
ekinv_high = data_high[7]#/data_high[8][0]
etot_high = data_high[8]/data_high[8][0]
max_dens_high = data_high[9]
a_high = data_high[10]

nstep_medium = data_medium[0]
time_medium = data_medium[1]
dt_medium = data_medium[2]
z_medium = data_medium[3]
mass_medium = data_medium[4]/data_medium[4][0]
epot_medium = data_medium[5]/data_medium[5][0]
ekinrho_medium = data_medium[6]/data_medium[6][0]
ekinv_medium = data_medium[7]#/data_medium[8][0]
etot_medium = data_medium[8]/data_medium[8][0]
max_dens_medium = data_medium[9]
a_medium = data_medium[10]

nstep_low = data_low[0]
time_low = data_low[1]
dt_low = data_low[2]
z_low = data_low[3]
mass_low = data_low[4]/data_low[4][0]
epot_low = data_low[5]/data_low[5][0]
ekinrho_low = data_low[6]/data_low[6][0]
ekinv_low = data_low[7]#/data_low[8][0]
etot_low = data_low[8]/data_low[8][0]
max_dens_low = data_low[9]
a_low = data_low[10]

hubl = 0.7
comoving_OmAx = 1.0
meandens = 2.775e+11 * hubl**2* comoving_OmAx
start = 0

line_x = np.linspace(0.03, 0.3, num=1000)
line_y = [(x*100.0) for x in line_x]

plt.loglog(a_high[start:],(max_dens_high[start:]-meandens)/(max_dens_high[start]-meandens),label='high')
plt.loglog(a_medium[start:],(max_dens_medium[start:]-meandens)/(max_dens_medium[start]-meandens),label='medium')
plt.loglog(a_low[start:],(max_dens_low[start:]-meandens)/(max_dens_low[start]-meandens),label='low')
plt.loglog(line_x,line_y,label='a^1')

plt.ylim([5e-2,5e+2])

plt.legend()
```
